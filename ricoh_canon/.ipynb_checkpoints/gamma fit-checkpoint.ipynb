```python
import csv
import numpy as np
import cv2
import plotly.graph_objects as go
import plotly.offline as po
import plotly.io as pio
import plotly
from scipy.optimize import curve_fit
import pprint
import openpyxl

gamma = 0
c = 0

def func(X,gamma,c):
    x = X
    return (c*((x/255)**gamma)*255)
          
    
x_demo=[]

y = np.loadtxt('plot.csv', delimiter=',', dtype='float')
    
for i in range(0,52):
    x_demo.append(i*5)

x = np.array(x_demo)

popt,pcov = curve_fit(func,x,y,[gamma,c])
k = func(x,popt[0],popt[1])

trace1 = plotly.graph_objs.Scatter(x=x, y=y,mode='markers', marker=dict(size=5, color="rgb(0,0,255)"))
trace2 = plotly.graph_objs.Scatter(x=x, y=k,mode='markers', marker=dict(size=5, color="rgb(255,0,0)"))

data = [trace1,trace2]
layout = plotly.graph_objs.Layout(height=600,width=800,title='psf')
figure = plotly.graph_objs.Figure(data=data, layout=layout)
po.plot(figure, filename="Pro2_Red_gamma.html")

print(popt[0])
print(popt[1])
```

```python
import torchvision.transforms.functional as TF
import lpips
from PIL import Image

loss_fn_alex = lpips.LPIPS(net='alex')

img_row = Image.open('a.png')
img_row = (TF.to_tensor(img_row))
img_row.unsqueeze(0)

img_hikaku = Image.open('b.png')
img_hikaku = (TF.to_tensor(img_hikaku))
img_hikaku.unsqueeze(0)

d = loss_fn_alex(img_row, img_hikaku)
print("Perceptual loss",d)
```

```python
import torchvision.transforms.functional as TF
import lpips
from PIL import Image

loss_fn_alex = lpips.LPIPS(net='alex')

for num in range(6):
    
    for num2 in range(10):
    
        if num == 0:
            filename_row = 'Mandrill_%d.jpg' % (10 - num2)
            filename_hikaku = 'Mandrill_hosei_clip_%d.jpg' % (10 - num2)
            
        if num == 1:
            filename_row = 'Parrots_%d.jpg' % (10 - num2)
            filename_hikaku = 'Parrots_hosei_clip_%d.jpg' % (10 - num2)
            
        if num == 2:
            filename_row = 'house_%d.jpg' % (10 - num2)
            filename_hikaku = 'house_hosei_clip_%d.jpg' % (10 - num2)
            
        if num == 3:
            filename_row = 'boart_%d.jpg' % (10 - num2)
            filename_hikaku = 'boart_hosei_clip_%d.jpg' % (10 - num2)
        
        if num == 4:
            filename_row = 'airplane_%d.jpg' % (10 - num2)
            filename_hikaku = 'airplane_hosei_clip_%d.jpg' % (10 - num2)
            
        if num == 5:
            filename_row = 'green_%d.jpg' % (10 - num2)
            filename_hikaku = 'green_hosei_clip_%d.jpg' % (10 - num2)

        img_row = Image.open(filename_row)
        img_row = (TF.to_tensor(img_row))
        img_row.unsqueeze(0)

        img_hikaku = Image.open(filename_hikaku)
        img_hikaku = (TF.to_tensor(img_hikaku))
        img_hikaku.unsqueeze(0)

        d = loss_fn_alex(img_row, img_hikaku)
        print(d)
        
    print("---------------------------------------")
```
